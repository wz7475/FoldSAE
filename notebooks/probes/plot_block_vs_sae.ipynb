```python
import matplotlib.pyplot as plt
import json

import numpy as np
```

```python
with open("data/block_activatons.json") as fp:
    block_activations = json.load(fp)
with open("data/latents_probes_3.json") as fp:
    sae_activations = json.load(fp)
with open("data/concat_block_act.json") as fp:
    concat_sae_activations = json.load(fp)
```

```python
fig, ax  = plt.subplots()
ax.invert_xaxis()
plt.plot(block_activations.values(), label="block activations (NON-PAIR)")
plt.plot(sae_activations.values(), label="SAE latents (NON-PAIR)")
plt.plot(concat_sae_activations.values(), label="block activations (NON-PAIR concat PAIR)")
plt.legend()
```
